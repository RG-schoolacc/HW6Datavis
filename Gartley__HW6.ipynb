b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'], which you may have to convert date type again, put those years as labels, and set the titles for each small plot to their state names. In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years.

```python
#your code goes here
g = sns.FacetGrid(filtered_df, col="state", col_wrap=3)
g.map(sns.lineplot, "date", "house_price_index", color="red")
tick_dates=pd.to_datetime(['1980', '2000', '2020'])
tick_labels = ['1980', '2000', '2020']
for ax in g.axes.flat:
    ax.set_xticks(tick_dates)
    ax.set_xticklabels(tick_labels)
    ax.set_xlabel('') # Remove x-axis label

# Set the y-axis label and individual plot titles
g.set_axis_labels("", "house price index")
g.set_titles("{col_name}")
```

```python
#Your code goes here.
world_cup = pd.read_csv('/content/drive/MyDrive/myDATA/WorldCupMatches.csv')
world_cup.head()
```

```python
#your code goes here. Just leaving an example output of an intermediate step for you to check the last two columns
#A hint to start with:
teams = sorted(['Home Team Initials', 'Away Team Initials'])
world_cup['Team1'] = world_cup[['Home Team Initials', 'Away Team Initials']].min(axis=1)
world_cup['Team2'] = world_cup[['Home Team Initials', 'Away Team Initials']].max(axis=1)
grouped = world_cup.groupby(['Team1', 'Team2']).agg(weight=('Home Team Initials', 'count'),HomeGoalTotal=('Home Team Goals', 'sum')).reset_index()
team_pairs = grouped.rename(columns={'Team1': 'Home Team Initials','Team2': 'Away Team Initials'})
```

Make sure that you only write four columns ('Home Team Initials', 'Away Team Initials', 'weight', 'HomeGoalTotal') from the data frame, team_pairs, to a csv file

```python
final_columns = ['Home Team Initials', 'Away Team Initials', 'weight', 'HomeGoalTotal']
team_pairs[final_columns].to_csv('team_pairs.csv', index=False)
```
